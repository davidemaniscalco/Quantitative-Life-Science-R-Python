```python
LL8 = np.loadtxt('./files_new_idea/LL_8.csv')
LL16 = np.loadtxt('./files_new_idea/LL_16.csv')
LL32 = np.loadtxt('./files_new_idea/LL_32.csv')
LL64 = np.loadtxt('./files_new_idea/LL_64.csv')
LL128 = np.loadtxt('./files_new_idea/LL_128.csv')
LL256 = np.loadtxt('./files_new_idea/LL_256.csv')
LL512 = np.loadtxt('./files_new_idea/LL_512.csv')

pp_list = np.arange(0.2,0.41,0.01)
pp_list_8 = np.arange(0.4,0.71,0.01)
```

```python
def Intercept_Plot(LL,pp_list,LL_array):
    coef, inter = np.polyfit(pp_list, np.log(np.mean(LL_array,axis=0)),1)
    xx = np.arange(np.min(pp_list),np.max(pp_list),0.01)
    p_star = (np.log(np.sqrt(3*np.pi*LL/2)) - inter)/coef

    plt.scatter(pp_list,np.mean(LL_array,axis=0),color='green',zorder=2)
    plt.axhline(np.sqrt(3*np.pi*LL/2),linestyle='--',color='black',zorder=2,label='$\\sqrt{3\pi L/2}$')
    plt.xlabel('$p$')
    plt.ylabel('$\\langle R \\rangle$')
    plt.title('L = '+str(LL))
    plt.plot(xx, np.exp(inter + xx*coef),color='lime',zorder=1)
    plt.scatter(p_star,np.sqrt(3*np.pi*LL/2),color='red',marker='*',s=60,zorder=3,
                label='$p^{*}\\approx \,$'+"%.2f" % p_star)
    plt.grid(alpha=.25)
    plt.plot()
    plt.legend()
    plt.show()
```
